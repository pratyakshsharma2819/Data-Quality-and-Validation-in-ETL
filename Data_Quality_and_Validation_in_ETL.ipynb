Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?

    ->Data quality in ETL pipelines refers to the overall fitness of data for its intended use, ensuring it meets business and analytical needs throughout extraction, transformation, and loading stages.
    Data cleaning addresses surface issues like nulls or formats, but quality involves proactive validation against business rules, metadata lineage tracking, and ongoing monitoring to prevent downstream errors in analytics

 Explain why poor data quality leads to misleading dashboards and incorrect decisions.
    ->Poor data quality in ETL pipelines undermines the reliability of dashboards and leads to flawed business decisions by introducing errors that cascade through analytics.
    Inaccurate, incomplete, or inconsistent data from ETL processes creates misleading visualizations, such as inflated sales metrics or skewed trend lines, as flawed inputs propagate to BI tools like Tableau or Power BI.

What is duplicate data? Explain three causes in ETL pipelines.

    ->Duplicate data refers to identical or near-identical records that appear multiple times within a dataset, violating the uniqueness dimension of data quality and leading to skewed analytics.
    Multiple Data Sources:When ETL extracts from overlapping sources like CRM and ERP systems, the same customer record may enter the pipeline twice without deduplication logic, creating redundant entries during integration.
    Faulty Extraction Logic:Rerunning extraction jobs without proper state tracking—such as missing incremental loads or checkpoints—re-ingests full datasets, duplicating prior records instead of capturing only changes.
    Improper Joins in Transformation:Mismatched join keys or Cartesian products during SQL merges generate unintended duplicates, especially with fuzzy matching failures on similar but non-identical values like name variations.

 Differentiate between exact, partial, and fuzzy duplicates.

    ->Exact duplicates match records identically across all key fields, like two rows with the same customer ID, name, and email. Partial duplicates share some but not all attributes, such as matching on name and city but differing in postal code due to data truncation. Fuzzy duplicates are approximate matches based on similarity algorithms

Why should data validation be performed during transformation rather than after loading?
    
    ->Data validation during ETL transformation catches transformation-specific errors—like faulty business logic or aggregations—before flawed data contaminates the data warehouse.

Explain how business rules help in validating data accuracy. Give an example.

    ->Business rules in data validation act as predefined logic reflecting organizational policies, ensuring data accuracy by cross-checking values against real-world constraints and dependencies beyond simple syntax.
    In a retail ETL pipeline, a business rule might state: "Discount percentage must be between 0% and 50%, and total sale price cannot exceed original price." If a transformed record shows a 75% discount making the sale price negative, validation rejects it, catching calculation errors early.

SELECT
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount,
    COUNT(*) as Duplicate_Count
FROM Sales_Transactions
GROUP BY Customer_ID, Product_ID, Txn_Date, Txn_Amount
HAVING COUNT(*) > 1
ORDER BY Duplicate_Count DESC;

-- Identify Sales_Transactions.Customer_ID values violating referential integrity
SELECT DISTINCT
    st.Customer_ID AS Violating_Customer_ID,
    COUNT(*) as Violation_Count
FROM Sales_Transactions st
LEFT JOIN Customers_Master cm
    ON st.Customer_ID = cm.Customer_ID
WHERE cm.Customer_ID IS NULL
GROUP BY st.Customer_ID
ORDER BY Violation_Count DESC;
